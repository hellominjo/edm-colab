```python
# ===============================================
# 2.2.2 나이에 따른 세대 구분 (1)
# 사용자의 출생 연도를 입력받아, 해당 세대를 한국어로 구분하여 출력하는 프로그램
# ===============================================

# 사용자의 출생 연도를 입력받음 (문자열 → 정수형 변환)
year = int(input("당신이 태어난 연도를 입력하세요: "))

# 입력된 연도에 따라 세대를 조건문으로 구분
if year <= 1924:
    generation = "가장 위대한 세대(The Greatest Generation)"
elif 1925 <= year <= 1945:
    generation = "침묵 세대(The Silent Generation)"
elif 1946 <= year <= 1964:
    generation = "베이비붐 세대(Baby Boomer)"
elif 1965 <= year <= 1980:
    generation = "X세대(Generation X)"
elif 1981 <= year <= 1996:
    generation = "밀레니얼 세대(Millennial)"
else:  # 1997년 이후 출생자
    generation = "Z세대(Generation Z)"

# 결과 출력
print(f"당신은 {generation}에 속합니다.")

# ===============================================
# 실행 결과 예시
# 당신이 태어난 연도를 입력하세요: 1946
# 당신은 베이비붐 세대(Baby Boomer)에 속합니다.
# ===============================================

# 배운 점
# - input()으로 입력받은 값은 기본적으로 문자열이므로 int()로 변환해야 숫자 비교 가능
# - if~elif~else 문으로 조건을 세분화하여 논리적으로 구분할 수 있음
# - f-string을 활용하면 변수값을 자연스럽게 문장 안에 삽입 가능
# - 코드의 가독성을 높이기 위해 들여쓰기와 주석을 명확히 하는 것이 중요함
```
